```python
# Subplot
fig3, sv = plt.subplots(5,3, figsize=(14,14))

# KDE Plot
b1 = sns.kdeplot(data=movies, x='BudgetMillions',y='CriticRating',cmap='Blues',ax=sv[0,0])
sv[0,0].set_title('KDE Plot - 01')
sv[0,0].set_xlabel('Budget in Millions')
sv[0,0].set_ylabel('Critic Rating')

b1 = sns.kdeplot(data=movies, x='BudgetMillions',y='CriticRating',cmap='Reds_r',ax=sv[0,1])
sv[0,1].set_title('KDE Plot - 02')
sv[0,1].set_xlabel('Budget in Millions')
sv[0,1].set_ylabel('Critic Rating')

b1 = sns.kdeplot(data=movies, x='BudgetMillions',y='CriticRating',cmap='Blues_r',shade= True, ax=sv[0,2])
sv[0,2].set_title('KDE Plot - 03')
sv[0,2].set_xlabel('Budget in Millions')
sv[0,2].set_ylabel('Critic Rating')

# SCATTER Plot
b2 = sns.scatterplot(data=movies, x='BudgetMillions',y='CriticRating',color = 'Yellow',ax=sv[1,0])
sv[1,0].set_title('SCATTER Plot - 01')
sv[1,0].set_xlabel('Budget in Millions')
sv[1,0].set_ylabel('Critic Rating')

b2 = sns.scatterplot(data=movies, x='BudgetMillions',y='CriticRating',color = 'Blue',ax=sv[1,1])
sv[1,1].set_title('SCATTER Plot - 02')
sv[1,1].set_xlabel('Budget in Millions')
sv[1,1].set_ylabel('Critic Rating')

b2 = sns.scatterplot(data=movies, x='BudgetMillions',y='CriticRating',color = 'Red',ax=sv[1,2])
sv[1,2].set_title('SCATTER Plot - 03')
sv[1,2].set_xlabel('Budget in Millions')
sv[1,2].set_ylabel('Critic Rating')

# HISTOGRAM
sv[2, 0].hist(movies['BudgetMillions'])
sv[2,0].set_title('HIST Plot - 01')
sv[2,0].set_xlabel('Budget in Millions')
sv[2,0].set_ylabel('Critic Rating')

sv[2, 1].hist(movies['CriticRating'])
sv[2,1].set_title('HIST Plot - 02')
sv[2,1].set_xlabel('Budget in Millions')
sv[2,1].set_ylabel('Critic Rating')

sv[2, 2].hist(movies['AudienceRating'])
sv[2,2].set_title('HIST Plot - 03')
sv[2,2].set_xlabel('Budget in Millions')
sv[2,2].set_ylabel('Critic Rating')

# VIOLIN Plot
b3 = sns.violinplot(data=movies, x='Genre', y='CriticRating', hue='Genre', ax=sv[3,0])
sv[3,0].set_title('VIOLIN Plot - 01')
sv[3,0].set_xlabel('Genre')
sv[3,0].set_ylabel('Critic Rating')

b3 = sns.violinplot(data=movies, x='Genre', y='AudienceRating', hue='Genre', ax=sv[3,1])
sv[3,1].set_title('VIOLIN Plot - 02')
sv[3,1].set_xlabel('Genre')
sv[3,1].set_ylabel('Audience Rating')

b3 = sns.violinplot(data=movies, x='Genre', y='BudgetMillions', hue='Genre', ax=sv[3,2])
sv[3,2].set_title('VIOLIN Plot - 03')
sv[3,2].set_xlabel('Genre')
sv[3,2].set_ylabel('Budget in Millions')


# BOX Plot
b4= sns.boxplot(data =movies[movies['Genre'] == 'Action'], x ='Year', y= 'BudgetMillions', color= 'red', ax=sv[4,0])
sv[4,0].set_title('Boxplot of Budget for Action Movies')
sv[4,0].set_xlabel('Year')
sv[4,0].set_ylabel('Budget in Millions')

b4= sns.boxplot(data =movies[movies['Genre'] == 'Adventure'], x ='Year', y= 'BudgetMillions', color= 'blue', ax=sv[4,1])
sv[4,1].set_title('Boxplot of Budget for Adventure Movies')
sv[4,1].set_xlabel('Year')
sv[4,1].set_ylabel('Budget in Millions')

b4= sns.boxplot(data =movies[movies['Genre'] == 'Drama'], x = 'Year', y= 'BudgetMillions', color= 'yellow', ax=sv[4,2])
sv[4,2].set_title('Boxplot of Budget for Drama Movies')
sv[4,2].set_xlabel('Year')
sv[4,2].set_ylabel('Budget in Millions')


# Map Layout Adjust
plt.tight_layout()

plt.show()
```
